```python
# 
train['Feature 10'] = train['Feature 10'].astype('category').cat.codes
train['Feature 11'] = train['Feature 11'].astype('category').cat.codes
train['Feature 12'] = train['Feature 12'].astype('category').cat.codes
```

```python
# Dropping since it has unique values
train = train.drop('Feature 13', axis=1)
```

```python
train = train.drop('Feature 14', axis=1)
```

```python
import pandas as pd
import numpy as np

from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
X = train.drop('Class',axis=1)
Y = train['Class']
Y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, random_state=42)
```

```python
train.corr()
```

```python
# import seaborn as sns
# sns.pairplot(data=train)
```

```python
train.describe
```

```python
from sklearn import preprocessing
```

```python
normal_X = preprocessing.scale(xtrain)
normal_Xtest = preprocessing.scale(xtest)
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
```

```python
lgbm = LGBMClassifier(num_leaves=70, objective='binary', random_state=5)

lgbm.fit(normal_X, ytrain)

y_pred = lgbm.predict(normal_Xtest)
```

```python
y_pred
```

```python
accuracy_score(ytest, y_pred)
```

```python
test = pd.read_csv('test.csv').drop('ID',axis=1)
```

```python
test = test.drop('Feature 14', axis=1)
```
